```python
#zad 6 wersja dosłowna

import numpy as np


A = np.array([[1, 1, 5],
             [2, 0, 6],
             [8, 3, 2]])

B = np.array([[3, -1, 1],
             [5, 1, 4],
             [-1, 3, 2]])

C = np.array([[1, 3, -2, 4],
             [1, -1, 3, 5],
             [0, 1, 4, -2],
             [10, -2, 5, 1]])

D = np.array([[2, 8, 3, -4],
             [1, 4, 1, -2],
             [5, 20, 0, -10],
             [-3, -12, -2, 6]])

E = np.array([[1,1],
             [1,1]])

F = np.array([[0]])


def get_det(A):
    if len(A[0])==2:
        return A[0,0] * A[1,1] - A[0,1] * A[1,0]
    
    if len(A[0])==3:
        return (A[0][0]* (A[1][1]*A[2][2] - A[1][2] * A[2][1]))-\
               (A[0][1]* (A[1][0]*A[2][2] - A[1][2]*A[2][0])) +\
                A[0][2]* (A[1][0]*A[2][1] - A[1][1]*A[2][0])
    
    if len(A[0])==4:
        return A[0][0]*A[1][1]*A[2][2]*A[3][3] + A[0][0]*A[1][2]*A[2][3]*A[3][1] + A[0][0]*A[1][3]*A[2][1]*A[3][2]+\
               A[0][1]*A[1][0]*A[2][3]*A[3][2] + A[0][1]*A[1][2]*A[2][0]*A[3][3] + A[0][1]*A[1][3]*A[2][2]*A[3][0]+\
               A[0][2]*A[1][0]*A[2][1]*A[3][3] + A[0][2]*A[1][1]*A[2][3]*A[3][0] + A[0][2]*A[1][3]*A[2][0]*A[3][1]+\
               A[0][3]*A[1][0]*A[2][2]*A[3][1] + A[0][3]*A[1][1]*A[2][0]*A[3][2] + A[0][3]*A[1][2]*A[2][1]*A[3][0]+\
              -A[0][0]*A[1][1]*A[2][3]*A[3][2] - A[0][0]*A[1][2]*A[2][1]*A[3][3] - A[0][0]*A[1][3]*A[2][2]*A[3][1]+\
              -A[0][1]*A[1][0]*A[2][2]*A[3][3] - A[0][1]*A[1][2]*A[2][3]*A[3][0] - A[0][1]*A[1][3]*A[2][0]*A[3][2]+\
              -A[0][2]*A[1][0]*A[2][3]*A[3][1] - A[0][2]*A[1][1]*A[2][0]*A[3][3] - A[0][2]*A[1][3]*A[2][1]*A[3][0]+\
              -A[0][3]*A[1][0]*A[2][1]*A[3][2] - A[0][3]*A[1][1]*A[2][2]*A[3][0] - A[0][3]*A[1][2]*A[2][0]*A[3][1]
    
    return '0'

def rank(A):
    if len(A[0])!=len(A):
        return 'Podaj macierz kwadratową'
    size = len(A[0])
    if size == 4:
        if get_det(A)!= 0:
            return '4'
        size -=1
    if size==3:
        if len(A[0])==4:
            for x in range(len(A[0])):
                for y in range(len(A)):
                    P = A.copy()
                    P = np.delete(P,x,0)
                    P = np.delete(P,y,1)
                    if get_det(P)!=0:
                        return '3'
        if len(A[0])==3:
            for x in range(len(A[0])):
                for y in range(len(A)):
                    P = A.copy()
                    if get_det(P)!=0:
                        return '3'
        size -= 1
    if size==2:
        if len(A[0])==4:
            for x in range(len(A[0])):
                for y in range(len(A)):
                    P = A.copy()
                    P = np.delete(P,x,0)
                    P = np.delete(P,x,0)
                    P = np.delete(P,y,1)
                    P = np.delete(P,y,1)
                    if get_det(P)!=0:
                        return '2'
        if len(A[0])==3:
            for x in range(len(A[0])):
                for y in range(len(A)):
                    P = A.copy()
                    P = np.delete(P,x,0)
                    P = np.delete(P,y,1)
                    if get_det(P)!=0:
                        return '2'
        if len(A[0])==2:
            for x in range(len(A[0])):
                for y in range(len(A)):
                    P = A.copy()
                    if get_det(P)!=0:
                        return '2'
        size -= 1
        
    if size ==1:
        for x in range(len(A[0])):
            for y in range(len(A)):
                if A[x][y]!=0:
                    return '1'
        
    return '0'

print("Ręcznie rząd a)",rank(A))
print("Z funkcji np a)",np.linalg.matrix_rank(A))
print("\nRęcznie rząd b)",rank(B))
print("Z funkcji np b)",np.linalg.matrix_rank(B))
print("\nRęcznie rząd c)",rank(C))
print("Z funkcji np c)",np.linalg.matrix_rank(C))
print("\nRęcznie rząd d)",rank(D))
print("Z funkcji np d)",np.linalg.matrix_rank(D))

print("\n\nTestowe przypadki\n\nRęcznie rząd e)",rank(E))
print("Z funkcji np e)",np.linalg.matrix_rank(E))
print("\nRęcznie rząd f)",rank(F))
print("Z funkcji np f)",np.linalg.matrix_rank(F))

print("\nWyznacznik ręcznie: ", get_det(C))
print("Wyznacznik z funkcji: ", np.linalg.det(C))
```
