```python
# zad 1

import numpy as np

A = np.array([[8, 1, 4, 2, 1],
             [8, 6, 4, 2, 1],
             [1, 2, 3, 4, 1],
             [8, 0, 6, 2, 5],
             [1, 9, 6, 4, 1]])

def inv(A):
    # tworzymy macierz jednostkową na bazie macierzy A
    id = np.eye(len(A))
    # wyznaczamy macierz rozszerzoną, będącą połączeniem macierzy A oraz odpowiadającej jej macierzy jednostkowej
    cA = np.concatenate((A, id), axis=1) 
    # przechodzimy przez wiersze macierzy                                     
    for x in range(len(A)): 
        # tworzymy zmienną odpowiedzialną za kolumny
        y = 1
        # wyznaczamy wartość przekątnej
        diag = cA[x][x] 
        # szukamy niezerowego współczynnika
        while x + y < len(A):
            # zamieniamy wiersze
            cA[[x, x + y]] = cA[[x + y, x]]
            # wyznaczanmy nowy współczynnik przekątnej
            diag = cA[x][x] 
            y += 1
        # bierzemy rząd i dzielimy w celu otrzymania 1 na przekątnej 
        cA[x] /= diag
        # przechodzimy przez rzędy omijając przekątną
        for y in [k for k in range(len(A)) if k != x]: 
            # odejmujemy obecny rząd od pozostałych
            cA[y] -= cA[x] * cA[y][x] 
    # zwracamy macierz odwrotna pozbywając się przeksztalconej macierzy jednostkowej
    return cA[:, len(A):] 

print("A*A^(-1)=\n", np.around(((inv(A)@A)),decimals=0))
print("Macierz odwrotna liczona ręcznie:\n", inv(A))
print("Macierz odwrotna liczona za pomocą np:\n", np.linalg.inv(A))
```
